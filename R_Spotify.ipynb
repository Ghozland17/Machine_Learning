# Machine Learning Projet : Prédiction de la popularité d'une bande sonore.



Les données extraites de Spotify Web API sont décrites par les variables suivantes :   
**valence** (the positiveness of the track),   
**year** (the release year of
track),   
**acousticness** (the relative metric of the track being acoustic),   
**danceability** (the relative measurement of the track being danceable),   
**duration** (the length of the track in milliseconds),   
**energy**(the energy of the track),   
**instrumentalness** (the relative ratio of the track being instrumental),   
**key**(the primary key of the track),  
**liveness**  (the relative duration of the track sounding as a live performance),  
**loudness** (the relative loudness of the track in the typical range [-60,0] in decibel),   
**mode** (a binary value representing whether the track starts with a major [encoded 1]
chord progression or not [encoded 0]),   
**speechiness** (the relative length of the track containing any
kind of human voice),   
**tempo** (the tempo of the track in Beat Per Minute).

```r
spotify <- read.table(paste("", "spotify-extr.txt", sep = ""),
                    sep = "", header = TRUE)
summary(spotify)
```

```r
spotify[, "key"] = as.factor(spotify[, "key"])
spotify[, "pop.class"] = as.factor(spotify[,"pop.class"])
```

```r
par(mfrow=c(3,3))
hist(spotify[,"valence"])
hist(spotify[,"year"])
hist(spotify[,"acousticness"])
hist(spotify[,"danceability"])
hist(spotify[,"duration"])
hist(spotify[,"energy"])
hist(spotify[,"instrumentalness"])
hist(spotify[,"liveness"])
hist(spotify[,"loudness"])
hist(spotify[,"mode"])
hist(spotify[,"popularity"])
hist(spotify[,"speechiness"])
hist(spotify[,"tempo"])




```

```r
spotify[,"Lspeechiness"]<- log(spotify[,"speechiness"])
spotify[,"Lduration"]<- log(spotify[,"duration"])
spotify[,"Lliveness"]<- log(spotify[,"liveness"])
spotify[,"Lpopularity"]<- log(spotify[,"popularity"])

par(mfrow=c(2,2))
hist(spotify[,"Lduration"])
hist(spotify[,"Lspeechiness"])
hist(spotify[,"Lliveness"])
hist(spotify[,"Lpopularity"])
```
